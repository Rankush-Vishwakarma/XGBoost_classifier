```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
lr = LogisticRegression(class_weight ='balanced', solver = 'lbfgs', random_state=42, max_iter = 500)
```

```python
X_train, X_test, y_train , y_test = train_test_split(X,y, test_size = 0.3)
```

```python
from sklearn.metrics import accuracy_score

# Init classifier
xgb_cl = xgb.XGBClassifier()

# Fit
xgb_cl.fit(X_train, y_train)

# Predict
preds = xgb_cl.predict(X_test)
```

```python
# accuracy 
accuracy_score(y_test, preds)
```

# hyper parameter tuning

```python
param_grid = {
    "max_depth": [3, 4, 5, 7],
    "learning_rate": [0.1, 0.01, 0.05],
    "gamma": [0, 0.25, 1],
    "reg_lambda": [0, 1, 10],
    "scale_pos_weight": [1, 3, 5],
    "subsample": [0.8],
    "colsample_bytree": [0.5],
}
```

```python
from sklearn.model_selection import GridSearchCV

# Init classifier
xgb_cl = xgb.XGBClassifier(objective="binary:logistic")

# Init Grid Search
grid_cv = GridSearchCV(xgb_cl, param_grid, n_jobs=-1, cv=3, scoring="roc_auc")

# Fit
_ = grid_cv.fit(X_train, y_train)
```

```python
 grid_cv.best_score_
```

```python
grid_cv.best_estimator_
```

```python
grid_cv.best_index_
```

```python
grid_cv.best_params_
```

```python
pred = grid_cv.predict(X_test)
```

```python
# accuracy 
accuracy_score(y_test, pred)
```

```python
import pickle
file_name = "xgb_classifier.pkl"

# save the normal one
pickle.dump(xgb_cl, open(file_name, "wb"))
file_name = "xgb_classifier_tuned.pkl"
# saving the hyper parameter tuned model
pickle.dump(grid_cv, open(file_name, "wb"))
```

```python
data = pd.read_csv('/content/drive/MyDrive/complete_data.csv')
```

```python
data.head()
```

```python
import numpy as np
np.unique(data['wage_class'])
```

```python
data['wage_class'].replace({' <=50K.': ' <=50K',' >50K.':' >50K' }, inplace = True)
```

```python
data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
data.head()
```

```python
np.unique(data['wage_class'])
```

```python
categorical_colunms = [i for i in data.columns if data[i].dtypes == 'O']
```

```python
data[categorical_colunms].head()
```

```python
for feature in categorical_colunms:
  d = dict(enumerate(data[feature].astype('category').cat.categories))
  print(f'feature-{feature}',d)
  data[feature] = data[feature].astype('category').cat.codes
```

```python
df1 = data[categorical_colunms]
```

We have categorical feature and categorical class so we can use 
    either Chi-Squared test (contingency tables) or Mutual Information.

```python
from sklearn.feature_selection import mutual_info_classif 
```

```python
X = df1.drop('wage_class', axis=1)
y = df1['wage_class']
```

```python
mi_score = mutual_info_classif(X,y)
print(mi_score, len(mi_score))
```

Let's try to choose the best top feature 

choosing top 5 feature from categroical variable

```python
print(sorted(range(len(mi_score)), key=lambda i: mi_score[i], reverse=True)[:5])
```

```python
cat_feature = [categorical_colunms[i] for i in [4, 2, 1, 3, 6]]
```

```python
cat_feature
```

WE have the top 5 categorical features now

Now let's do the same for numerical columns

```python
data.head()
```

```python
data.drop('education_num', axis=1,inplace=True)
```

```python
numerical_feature =  [feature for feature in data.columns if data[feature].dtype != 'O' and feature not in categorical_colunms]
```

```python
data[numerical_feature+['wage_class']]
```

```python
for column in numerical_feature:
    data[column] = data[column]  / data[column].abs().max()
```

```python
df2 = data[numerical_feature+['wage_class']]
```

```python
X2 = df2.drop('wage_class', axis=1)
y2 = df2['wage_class']
```

WE have numerical input and categorical output for this we can use 
 either ANOVA correlation coefficient (linear) or Kendallâ€™s rank coefficient (nonlinear).

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
fvalue_Best = SelectKBest(f_classif, k=3)
X_kbest = fvalue_Best.fit_transform(X2, y2)
print(X_kbest)
```
